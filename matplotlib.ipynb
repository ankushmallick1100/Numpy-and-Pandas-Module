```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,2,3]
y1 = [7,8,9]
x2 = [14,15,19]
y2 = [20,25,27]
plt.plot(x1,y1, label='first',linewidth=2)
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,2,3]
y1 = [7,8,9]
x2 = [14,15,19]
y2 = [20,25,27]
plt.scatter(x1,y1, label='first',linewidth=2)
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x1 = [1,2,3]
y1 = [7,8,9]
x2 = [14,15,19]
y2 = [20,25,27]
plt.hist(x1, label='first',linewidth=2)
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```
